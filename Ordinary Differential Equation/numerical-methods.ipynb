```python
import numpy as np
import matplotlib.pyplot as plt
```

Calculating numerical and analytical solution for a given differential equation and seeing how they differ from each other.

```python
#1

a=0                             #given limits
b=1
lamda=1                        #constant
N=10
x = np.linspace(a,b,N)       #dividing the interval into N segments
delx = (b-a)/N                 #the small segments

#Numerical_Solution
y=[1]
for i in range(1,N):
    yn = (-lamda*delx + 1)*y[-1]
    y.append(yn)

#Analytic_solution
ya= np.exp(-x*lamda)

#plotting
plt.plot(x,y)
plt.plot(x,ya)
plt.legend(['Numerical','Analytical'])
plt.xlabel('x')
plt.ylabel('y')

#error
err = (np.sum(np.abs(y-ya)))/N
print('Error:',err)
```

Now we will see how the error changes when we increase/decrease the number of segments, i.e. N

```python
#First we will define a function which will calculate error for any N we call
#the code is same as above with just N removed, since we want error for different Ns.
def error(N):
    a=0
    b=1
    lamda=1
    x = np.linspace(a,b,N)
    delx=(b-a)/N
    y=[1]        #Numerical
    for i in range(1,N):
        yn=(-lamda*delx + 1)*y[-1]
        y.append(yn)
    ya= np.exp(-lamda*x)    #Analytical
    err=(np.sum(abs(y-ya)))/N  #error_between_them(Mean Global Error)
    return err
```

```python
NN= range(10,10000,5) #different number of Ns to calculate error
errors=[]   #an empty array to store all the errors corresponding to different Ns (starting from 10,15,..)
for i in NN:
    errors.append(error(i)) #changing the subinterval size to see how error varies
plt.xscale('log')  #using this because log scale makes it easy to compare values that cover a large range
plt.yscale('log')
plt.plot(NN,errors)
plt.xlabel('No. of sub-intervals(in log scale)')
plt.ylabel('Error (in log scale)')
```

```python
#2
#calculating slope of the above graph

(np.log(errors[20])-np.log(errors[12]))/(np.log(NN[20])-np.log(NN[12]))
```

Output discussion: Above we get a negative slope which indicates that N is inversely proportional to stepsize , which is directly proportional to error

Runge Kutta method

```python
#RK_2

a=0
b=1
N=10
lamda=5
x = np.linspace(a,b,N)
delx = (b-a)/N
def f(y):
    return -lamda*y
yn=[1]
for i in range(1,N):
    k1 = delx*f(yn[-1])
    k2=delx*f(yn[-1]+k1)
    y= yn[-1] + k1*(1/2) +k2*(1/2)
    yn.append(y)
    
ya=np.exp(-x*lamda)

plt.plot(x,yn)
plt.plot(x,ya)
```

```python
#RK_4

a=0
b=1
N=10
lamda=5
x = np.linspace(a,b,N)
delx = (b-a)/N
def f(y):
    return -lamda*y
yn=[1]
for i in range(1,N):
    k1 = delx*f(yn[-1])
    k2=delx*f(yn[-1]+k1/2)
    k3=delx*f(yn[-1]+k2/2)
    k4=delx*f(yn[-1]+k3)
    y= yn[-1] + 1/6*(k1 +2*k2+2*k3+k4)
    yn.append(y)
    
ya=np.exp(-x*lamda)

plt.plot(x,yn)
plt.plot(x,ya)
```
