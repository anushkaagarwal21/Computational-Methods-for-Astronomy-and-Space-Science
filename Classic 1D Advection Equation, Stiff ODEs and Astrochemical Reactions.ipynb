# Question 1
##Solving the equation
\begin{equation}
    \frac{du}{dt} = 5y - 5t^2
\end{equation}
## with initial condition
\begin{equation}
    y(0)=2
    \end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt    #necessary_libraries

#defining_given_function
def f(t,y):
    return 5*y - 5*t**2

n= [25,250,2500]    #different_no._of _grids_to_be_taken

for k in range(len(n)):    #Main_for_loop(which takes different number of grids each time)

    t = np.linspace(0,2,n[k])   #time_with_different_no._of_grids_each_method
    delt = t[1] - t[0]         #length_between_2_consecutive_grid

    y_ana = t**2 + (2/5)*t + (2/25)  #analytic_solution

    #initial_condition
    y_E = [2/25]
    y_RK2 = [2/25]
    y_BE = [2/25]

    #iterating_for_different methods
    for i in range(1,n[k]):
        #Euler
        y_E.append(y_E[-1] + delt * f(t[i],y_E[-1]))

        #Runge-Kutta2
        k1 = delt * f(t[i] , y_RK2[-1])
        k2 = delt * f(t[i]+delt , y_RK2[-1]+k1)
        y_RK2.append(y_RK2[-1] + k1/2 +k2/2)

        #Backward_Euler
        y_BE.append((y_BE[-1] - 5*delt*t[i]**2) / (1 - 5*delt) )

    fig, axs = plt.subplots(1,3,figsize=[20,4]) #creates_a_figure_with_three_subplots

    #PLOTTING FOR EACH METHOD
    axs[0].plot(t,y_E,label='Euler')
    axs[1].plot(t,y_RK2,label='Runge-Kutta 2')
    axs[2].plot(t,y_BE,label='Backward Euler')

    methods= ['Euler','RK-2','Backward Euler'] #different methods

    for s in range(len(methods)):
      axs[s].plot(t,y_ana,label='Analytical Solution')
      axs[s].set_title(f"Method:{methods[s]} for n:{n[k]}")
      axs[s].set_xlabel('t')
      axs[s].set_ylabel('y(t)')
      axs[s].grid(True,which='both')
      axs[s].legend()

plt.tight_layout() #to_prevent_overlap
plt.show()
```

# Question 2
### Solving the equation
\begin{equation}
    \frac{\partial u}{\partial t} + \frac{\partial u}{\partial x}= 0
\end{equation}



```python
import numpy as np
import matplotlib.pyplot as plt

# Defining parameters
nx = 200  # Number of grid points
x = np.linspace(-10,10,nx)
dx = x[1]-x[0]
nt = 1000  # Number of time steps
t = np.linspace(0,10,nt)
dt = t[1]-t[0]  # Time step
```

```python
lamda = 3   # Advection velocity

# Initial conditions
u0 = np.sqrt((1/2)*(np.pi)) * np.exp(-x**2/2)

# Plotting initial condition
plt.plot(x, u0, label='Initial Condition')

# Solving advection equation using finite difference method with periodic boundary conditions
# Copying the initial conditions for the array size and later updating them based on the function
u = u0.copy()
for n in range(nt):
    un = u.copy()
    for i in range(nx):
        u[i] = un[i] - lamda * dt / dx * (un[i] - un[(i - 1) % nx])

    # Plotting the solution for given time "instant"
    # We are using the updated u here
    plt.plot(x, u, label='Solution')
    plt.xlabel('x')
    plt.ylabel('u')
    plt.title('1D Advection Equation for Upwind scheme with lamda=3')
    plt.pause(0.08)
    plt.clf()
```

### CFL Condition :
The CFL (Courant Friedrichs Lewy) condition is a stability criteria, it ensures that we give enough time for the information to reach another grid. In other words, it ensures that information in the system doesn't propagate too quickly which lead to numerical instability.

\begin{equation}
CFL = \frac{ t}{x} \leq 1 \ \ \ (For \ 1D)
\end{equation}
